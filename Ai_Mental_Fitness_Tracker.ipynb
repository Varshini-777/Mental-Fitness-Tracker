# Split Data(6840,10)


*   In this step we are going to split data in 2 parts(training and testing)so that we can train our model on training dataset and test its accuracy on unseen test data.



```python
x=data.drop('mental_fitness',axis=1)
y=data['mental_fitness']
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.20,random_state=2)
```

```python
print("xtrain:",xtrain.shape)
print("xtest:",xtest.shape)
print("\n ytain: ",ytrain.shape)
print("ytest: ",ytest.shape)
```

**Takeaway Points**


*  Ususally we take more and more data in training so is easy for the model to learn with more data



# Model Training



* As we done with preprocessing part,its time to train our model.Here we are going to model our data with linear regression and random forest regressor algorithms and we will compare the both models and the choose the efficient one.



* Linear Regression is a popular and simple machine learning algorithm
used for regression tasks. It's used to model the relationship between a dependent variable (target) and one or more independent variables (features) by fitting a linear equation to the observed data.


* The equation for a simple linear regression with one feature is:



* y = b0 + b1 * x

Where:

y is the dependent variable (target).
x is the independent variable (feature).
b0 is the intercept (the value of y when x is 0).
b1 is the coefficient (the change in y for a one-unit change in x).







```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

lr = LinearRegression()
lr.fit(xtrain, ytrain)  # Instead of lr=fit(xtrain, ytrain)

ytrain_pred = lr.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = np.sqrt(mse)  # Use the calculated MSE for RMSE
r2 = r2_score(ytrain, ytrain_pred)  # Fix typo 'ytain_pred' to 'ytrain_pred'

print("The Linear Regression model performance for the training set")
print("-------------------------------------")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2 SCORE is {}".format(r2))




```



* Random Forest Regressor is a powerful ensemble learning algorithm used for regression tasks. It is an extension of the Random Forest algorithm, which is primarily used for classification tasks. The Random Forest Regressor is capable of handling both numerical and categorical features, and it is less prone to overfitting compared to a single decision tree.

*  It is used to solve both Regression and Classification problems.







```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(xtrain,ytrain)
ytrain_pred=rf.predict(xtrain)
mse=mean_squared_error(ytrain,ytrain_pred)
rmse=(np.sqrt(mean_squared_error(ytrain,ytrain_pred)))
r2=r2_score(ytrain,ytrain_pred)
print("The Random Forest Regressor model performance for training set")
print("-----------------------------------------------")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2SCORE is {}".format(r2))
```

# Evaluation



*  In this part we will see the scores of the both models.



```python
ytest_pred=lr.predict(xtest)
mse=mean_squared_error(ytest,ytest_pred)
rmse=(np.sqrt(mean_squared_error(ytest,ytest_pred)))
r2=r2_score(ytest,ytest_pred)
print("The Linear Regression modelperformance for testing set")
print("-------------------------------------")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2SCORE is {}".format(r2))


ytest_pred=rf.predict(xtest)
mse=mean_squared_error(ytest,ytest_pred)
rmse=(np.sqrt(mean_squared_error(ytest,ytest_pred)))
r2=r2_score(ytest,ytest_pred)
print("The Random Forest Regressor modelperformance for testing set")
print("-------------------------------------")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2SCORE is {}".format(r2))

```

Random Forest Regressor performs well on both training and testing data.
